```python
headers = {"Authorization": f"Bearer {API_TOKEN}"}
```

```python
def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()
```

```python
question = {'inputs':'What is capital of India?'}
output = query(question)
```

```python
print(output[0]['generated_text'])
```

# OpenAI API

Steps to get OpenAI API key:
1. Login/Signup to OpenAI website [Click Here](https://openai.com/index/openai-api/)
2. Once logged in, all documentation can be found [here](https://platform.openai.com/docs/overview)
3. Before creating the key you need to add the credits against you account. Click here to add the [credits](https://platform.openai.com/settings/organization/billing/overview).
4. Go the [following link](https://platform.openai.com/settings/profile?tab=api-keys) and click on "create new secret key".
5. Keep the key value safe as you cannot view it again.




```python
from google.colab import userdata
OPENAI_KEY = userdata.get('OPENAI_API_KEY')
```

Model selection from OPENAI

| Use Case | Model Name|
|---|---|
|text to text | GPT 3.5, GPT-3.5 turbo|
|text to image | DALL-E|
|Image to text | GPT 4 |
|Image/text to image/video| SORA |


```python
from openai import OpenAI
import pandas as pd
client = OpenAI(api_key = OPENAI_KEY)
models = client.models.list().data


# creating dataframe of models
models_df = pd.DataFrame(models, columns = ['id', 'created', 'object', 'owned_by'])
```

```python
# Generation of text

completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "YOu are a funny assistant who describe every thing in a funniest way"}, # Describes the behaviour of model
        {'role':'user', 'content': 'compose a poem for indian people' } # Question asked from the model
    ]
)

```

```python
print(completion.choices[0].message.content)
```

```python
# Using DALL-E

response = client.images.generate(
  model="dall-e-3",
  prompt="a white siamese cat",
  size="1024x1024",
  quality="standard",
  n=1,
)
```

```python
image_url = response.data[0].url
image_url
```

```python
import google.generativeai as genai
```

```python

from google.colab import userdata
GeminiAPI_key = userdata.get('GeminiAPI_key')
```

```python
genai.configure(api_key = GeminiAPI_key)

# To get all the model
for model in genai.list_models():
    print( model.supported_generation_methods , model.name)
```

```python
model = genai.GenerativeModel('gemini-1.0-pro')
response = model.generate_content('What is capital of india and give its small description')
```
